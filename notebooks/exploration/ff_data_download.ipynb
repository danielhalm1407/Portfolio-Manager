#### 0.2) Download Market Data (IBKR)

Run this cell once before anything else. Fetches 1Y of daily OHLCV data for all
portfolio tickers from IBKR, writes CSVs to `data/raw/market/`, and sets the
kernel working directory there so every downstream cell can load files with a
plain `pd.read_csv("AAPL.csv")`.

> **Requires** TWS or IB Gateway running on `127.0.0.1:7497` (paper or live).

```python


# ── Paths ───────────────────────────────────────────────────────────────────
# VS Code starts the kernel with cwd = sandbox/, so ".." resolves to project root.
# _MARKET_DIR is resolved to an absolute path BEFORE os.chdir() is called,
# so it stays valid even after the working directory is changed below.
_MARKET_DIR = (pathlib.Path.cwd() / ".." / ".." / "data" / "raw" / "market").resolve()
_MARKET_DIR.mkdir(parents=True, exist_ok=True)

# ── Tickers ─────────────────────────────────────────────────────────────────
# Defined inline here because sector_map is declared later in the notebook.
# SPY is included as a market benchmark (used in several downstream cells).
TICKERS = ["SPY",
            "AAPL", "MSFT", "AVGO",
            "UNH", "JNJ", "AMGN",
            "WMT", "COST", "PG",
            "CVX", "XOM", "LNG"]

# ── Download from IBKR ──────────────────────────────────────────────────────
print(f"Fetching up to {len(TICKERS)} tickers from IBKR...")
#print(f"Output  → {_MARKET_DIR}")
data = get_equity_data(
    symbols=TICKERS,
    duration="1 Y",
    bar_size="1 day",
    output_dir=_MARKET_DIR,   # saves stock_data_{sym}.csv there
)

# ── Re-save as bare {ticker}.csv ────────────────────────────────────────────
# Downstream cells load files as pd.read_csv("AAPL.csv"), so we write an
# additional copy without the stock_data_ prefix.
for sym, df in data.items():
    df.to_csv(_MARKET_DIR / f"{sym}.csv", index=False)
    print(f"  saved {sym}.csv  ({len(df)} rows)")

# ── Set cwd → data/raw/market/ ──────────────────────────────────────────────
# Every downstream cell uses bare filenames like pd.read_csv("AAPL.csv").
# Setting cwd here means none of those cells need to change.
os.chdir(_MARKET_DIR)
#print(f"Done. Working directory is now: {pathlib.Path.cwd()}")
```
